Импортируем библиотеку pandas. Загрузим датасет bookings.csv с разделителем ;. Проверим размер таблицы, типы переменных, а затем выведем первые 7 строк, чтобы посмотреть на данные. 

```python
import pandas as pd
bookings = pd.read_csv('datasets/bookings.csv', sep=';')
bookings.head(7)
```

```python
bookings.shape
```

```python
bookings.dtypes
```

Приведём названия колонок к нижнему регистру и заменим пробелы на знак нижнего подчеркивания.

```python
def rename_columns(name):
    name = name.lower().replace(' ', '_')
    return name

bookings = bookings.rename(columns=rename_columns)
bookings.head()
```

Выясним, пользователи из каких стран совершили наибольшее число успешных бронирований (топ-5).

```python
successful_bookings = bookings.query('is_canceled != 0')
successful_bookings.country.value_counts().head(5)
```

Выясним, на сколько ночей в среднем бронируют отели разных типов

```python
nights = bookings.groupby('hotel', as_index=False).agg({'stays_total_nights': 'mean'}).round(decimals=2)
nights
```

Иногда тип номера, полученного клиентом (assigned_room_type), отличается от изначально забронированного (reserved_room_type).  
Такое может произойти, например, по причине овербукинга. Посмотрим, сколько подобных наблюдений встретилось в датасете.

```python
overbooking = bookings.query('assigned_room_type != reserved_room_type')
overbooking.shape[0]
```

Выясним, на какой месяц чаще всего оформляли бронь в 2016 году и изменился ли самый популярный месяц в 2017.  

```python
popular_month_2016 = bookings.query('arrival_date_year == 2016').arrival_date_month.value_counts()
popular_month_2016
```

```python
popular_month_2017 = bookings.query('arrival_date_year == 2017').arrival_date_month.value_counts()
popular_month_2017
```

Сгруппируем данные по годам и проверим, на какой месяц бронирования отели типа City Hotel отменялись чаще всего в каждый из периодов.

```python
cancel_count = bookings.query('is_canceled == 1 and hotel == "City Hotel"').groupby(['arrival_date_year', 'arrival_date_month'], as_index=False).agg({'is_canceled': 'count'}).sort_values(['arrival_date_year', 'is_canceled'], ascending=[True, False])
cancel_count
```

Посмотрим на числовые характеристики трёх переменных: adults, children и babies и выясним, какая из них имеет наибольшее среднее значение.

```python
avg_babies = bookings.babies.mean()
avg_babies
```

```python
avg_children = bookings.children.mean()
avg_children
```

```python
avg_adults = bookings.adults.mean()
avg_adults
```

Создадим колонку total_kids, объединив children и babies и посмотрим, для отелей какого типа среднее значение переменной оказалось наибольшим.

```python
bookings_plus = bookings
bookings_plus['total_kids'] = bookings_plus.babies + bookings_plus.children
avg_kids = bookings_plus.groupby('hotel').agg({'total_kids': 'mean'}).round(decimals=2)
avg_kids
```

Создадим переменную has_kids, которая принимает значение True, если клиент при бронировании указал хотя бы одного ребенка (total_kids),  
в противном случае – False. Посчитаем отношение количества ушедших пользователей к общему количеству клиентов,  
выраженное в процентах (churn rate).

```python
bookings_plus['has_kids'] = bookings_plus['total_kids'] > 0
bookings_plus
```

```python
kids_cancel_rate = bookings_plus.query('has_kids == True').is_canceled.sum() / bookings_plus.query('has_kids == True').shape[0] * 100
kids_cancel_rate.round(decimals=2)
```

```python
nokids_cancel_rate = bookings_plus.query('has_kids == False').is_canceled.sum() / bookings_plus.query('has_kids == False').shape[0] * 100
nokids_cancel_rate.round(decimals=2)
```

Процент клиентов, которые отменили бронирование, выше среди тех, кто не указал детей при оформлении.
